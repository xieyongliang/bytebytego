```python
import pandas as pd
import numpy as np
```

```python
posts = {
    'post_id': [1, 2,3 ],
    'author_id': [1, 11, 4],
    'on_device': ['73.93.220.240', '89.42.110.250', '39.55.180.020'],
    'timestamp': [1658469431, 1658471428, 1658489233],
    'texual_content': ['Today, I am starting my diet.', 'The video amazed me! Please donate', 'What is a good restaurant in the Bay area?'],
    'images_or_videos': ['http: //cdn.mysite.com/u1.jpg', 'http: //cdn.mysite.com/t3.mp4', 'http: //cdn.mysite.com/t5.jpg'],
    'links': [np.nan, 'gofundme.com/f/3u1njd32', np.nan]
}
df_posts = pd.DataFrame(posts)
```

```python
df_posts['timestamp'] = pd.to_datetime(df_posts['timestamp'], unit='s')
```

```python
users = {
    'id': [1, 11, 4],
    'username': ['u1', 'u2', 'u3'],
    'age': [20, 22, 24],
    'gender': ['F', 'M', 'F'],
    'city': ['Beijing', 'Tianjin', 'Shanghai'],
    'country': ['China', 'Japan', 'USA'],
    'email': ['u1@a.b', 'u2@a.b', 'u3@a.b']
}
df_users = pd.DataFrame(users)
```

```python
df_users.gender = df_users.gender.astype('category')
df_users.city = df_users.city.astype('category')
df_users.country = df_users.country.astype('category')
df_users.dtypes
```

```python
df_posts_users = pd.merge(df_posts, df_users, left_on='author_id', right_on='id', how='left')
```

```python
df_posts_users.drop(['id'], axis=1)
```

```python
pd.get_dummies(df_posts_users, columns=['gender', 'city', 'country'])
```

```python
reactions = {
    'user_id': [11, 4, 11, 4, 11],
    'post_id': [6, 20, 7, 20, 7],
    'interaction_type': ['impression', 'like', 'comment', 'share', 'report'],
    'interaction_value': [np.nan, np.nan, 'this is disguesting', np.nan, 'violence'],
    'timestamp': [1658450539, 1658451341, 1658451365, 1658435948, 1658451849]
}
```

```python
df_reactions = pd.DataFrame(reactions)
df_reactions.interaction_type = df_reactions.interaction_type.astype('category')
df_reactions.timestamp = pd.to_datetime(df_reactions.timestamp, unit='s')
```

```python
df_posts_users_reactions = pd.merge(df_posts_users, df_reactions, on='post_id', how='left')
```

```python
df_reactions = pd.get_dummies(df_reactions, columns=['interaction_type'])
```

```python
bins = [0, 6, 12, 18, 24]
labels = ['late night', 'morning', 'afternoon', 'evening']

df_reactions['time_of_day'] = df_reactions.timestamp.dt.hour
```

```python
df_reactions['time_bucket'] = pd.cut(df_reactions['time_of_day'], bins=bins, labels=labels, right=False)
```

```python
time_bucket_dymmies = pd.get_dummies(df_reactions['time_bucket'], prefix='time')
```

```python
df_reactions_new = pd.concat([df_reactions, time_bucket_dymmies], axis=1)
```

```python
df_reactions_new.drop(columns=['time_of_day', 'time_bucket'])
```
